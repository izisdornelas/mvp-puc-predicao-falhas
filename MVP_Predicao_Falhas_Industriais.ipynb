# 6. Modelagem

Nesta etapa serão treinados e comparados três modelos de classificação supervisionada:

- Regressão Logística, utilizada como modelo baseline;
- Random Forest, como primeiro modelo candidato;
- Gradient Boosting, como segundo modelo candidato.

## 6.1 Regressão Logística


A Regressão Logística foi utilizada como modelo baseline deste projeto.

Esse algoritmo foi escolhido por ser uma abordagem simples, interpretável e amplamente utilizada em problemas de classificação binária. Seu desempenho servirá como referência para avaliar se modelos mais complexos conseguem obter resultados superiores na predição de falhas operacionais.

```python
# Importação das bibliotecas utilizadas no pré-processamento e treinamento
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

# Definição das variáveis categóricas e numéricas
categorical_features = ["Type"]

numeric_features = [
    "Air temperature [K]",
    "Process temperature [K]",
    "Rotational speed [rpm]",
    "Torque [Nm]",
    "Tool wear [min]"
]

# Construção do pré-processamento para variáveis numéricas e categóricas
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ]
)

# Construção da Pipeline utilizando Regressão Logística como modelo baseline
baseline_model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("classifier", LogisticRegression(max_iter=1000, random_state=42))
    ]
)

# Treinamento do modelo
baseline_model.fit(X_train, y_train)

print("Modelo treinado com sucesso.")

# Realização das previsões
y_pred_baseline = baseline_model.predict(X_test)
```

```python
y_pred_baseline = baseline_model.predict(X_test)

y_pred_baseline[:10]
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

# Cálculo das métricas de desempenho do modelo baseline
baseline_accuracy = accuracy_score(y_test, y_pred_baseline)
baseline_precision = precision_score(y_test, y_pred_baseline)
baseline_recall = recall_score(y_test, y_pred_baseline)
baseline_f1 = f1_score(y_test, y_pred_baseline)

print("Accuracy:", baseline_accuracy)
print("Precision:", baseline_precision)
print("Recall:", baseline_recall)
print("F1-score:", baseline_f1)
```

### Análise dos Resultados da Regressão Logística

O modelo baseline apresentou acurácia de aproximadamente 96,75%, valor aparentemente elevado.

Entretanto, devido ao forte desbalanceamento da variável-alvo, a acurácia isoladamente pode ser enganosa. O Recall foi de aproximadamente 10,3%, indicando que o modelo identificou apenas uma pequena parte das falhas reais.

Esse resultado mostra que a Regressão Logística possui limitação para detectar eventos raros de falha neste conjunto de dados.

## 6.2 Random Forest


O Random Forest foi escolhido como primeiro modelo candidato por ser um algoritmo baseado em árvores de decisão, capaz de capturar relações não lineares entre as variáveis.

Esse modelo costuma apresentar bom desempenho em dados tabulares e será comparado ao baseline para verificar possíveis ganhos na identificação de falhas.

```python
from sklearn.ensemble import RandomForestClassifier

# Construção da Pipeline utilizando o algoritmo Random Forest
rf_model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("classifier", RandomForestClassifier(random_state=42))
    ]
)

# Treinamento do modelo
rf_model.fit(X_train, y_train)

# Realização das previsões
y_pred_rf = rf_model.predict(X_test)
```

```python
rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_precision = precision_score(y_test, y_pred_rf)
rf_recall = recall_score(y_test, y_pred_rf)
rf_f1 = f1_score(y_test, y_pred_rf)

print("Accuracy :", rf_accuracy)
print("Precision:", rf_precision)
print("Recall   :", rf_recall)
print("F1-score :", rf_f1)
```

### Análise dos Resultados do Random Forest

O Random Forest apresentou desempenho superior ao baseline em todas as métricas avaliadas.

O principal ganho foi observado no Recall, que aumentou de aproximadamente 10,3% para cerca de 48,5%. Isso indica que o modelo conseguiu identificar uma quantidade maior de falhas reais.

O F1-score também apresentou melhora expressiva, mostrando melhor equilíbrio entre precisão e recall.

## 6.3 Gradient Boosting



O Gradient Boosting foi escolhido como segundo modelo candidato por ser um algoritmo de ensemble que constrói árvores de forma sequencial, corrigindo erros dos modelos anteriores.

Essa abordagem costuma apresentar bom desempenho em problemas de classificação com dados tabulares e será comparada aos modelos anteriores.

```python
from sklearn.ensemble import GradientBoostingClassifier

# Construção da Pipeline utilizando o algoritmo Gradient Boosting
gb_model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("classifier", GradientBoostingClassifier(random_state=42))
    ]
)

# Treinamento do modelo
gb_model.fit(X_train, y_train)

# Realização das previsões
y_pred_gb = gb_model.predict(X_test)
```

```python
gb_accuracy = accuracy_score(y_test, y_pred_gb)
gb_precision = precision_score(y_test, y_pred_gb)
gb_recall = recall_score(y_test, y_pred_gb)
gb_f1 = f1_score(y_test, y_pred_gb)

print("Accuracy :", gb_accuracy)
print("Precision:", gb_precision)
print("Recall   :", gb_recall)
print("F1-score :", gb_f1)
```

### Análise dos Resultados do Gradient Boosting

O Gradient Boosting apresentou o melhor desempenho entre os modelos avaliados.

O modelo alcançou Accuracy de aproximadamente 98,55%, Recall de aproximadamente 66,18% e F1-score de aproximadamente 75,63%.

Esses resultados indicam que o Gradient Boosting foi o modelo mais eficiente para identificar falhas operacionais neste MVP.

# 8. Otimização de Hiperparâmetros

Após a comparação dos modelos, o algoritmo Gradient Boosting foi selecionado para a etapa de otimização de hiperparâmetros, pois apresentou o melhor desempenho geral entre os modelos avaliados.

Foi utilizado o método GridSearchCV com validação cruzada para testar diferentes combinações de hiperparâmetros e identificar a configuração mais adequada para o problema proposto.

Os hiperparâmetros avaliados foram:

- n_estimators: quantidade de árvores utilizadas pelo modelo;
- learning_rate: taxa de aprendizado utilizada durante o treinamento;
- max_depth: profundidade máxima das árvores de decisão.

O critério de seleção adotado foi o F1-score, por ser uma métrica mais adequada para problemas com classes desbalanceadas, considerando simultaneamente Precision e Recall.

Mesmo sem ganhos expressivos em relação ao modelo inicial, a etapa de otimização foi mantida por fazer parte do fluxo recomendado de desenvolvimento de modelos de Machine Learning.

A validação cruzada permitiu avaliar diferentes combinações de parâmetros de forma mais robusta, reduzindo a dependência de uma única divisão dos dados.

```python
from sklearn.model_selection import GridSearchCV

# Definição do espaço de busca dos hiperparâmetros
param_grid = {
    "classifier__n_estimators": [50, 100],
    "classifier__learning_rate": [0.05, 0.1],
    "classifier__max_depth": [3, 5]
}

# Busca da melhor combinação de hiperparâmetros utilizando validação cruzada
grid_search = GridSearchCV(
    estimator=gb_model,
    param_grid=param_grid,
    scoring="f1",
    cv=3,
    n_jobs=-1
)

# Execução da busca pelos melhores hiperparâmetros
grid_search.fit(X_train, y_train);
```

```python
print("Melhores parâmetros:")
print(grid_search.best_params_)

print("\nMelhor F1 na validação cruzada:")
print(grid_search.best_score_)
```

## 8.1 Resultados da Otimização

A otimização identificou como melhor configuração os seguintes hiperparâmetros:

- learning_rate = 0,1
- max_depth = 5
- n_estimators = 50

O melhor F1-score obtido durante a validação cruzada foi de aproximadamente 0,715.

Esses resultados indicam que a combinação selecionada apresentou o melhor equilíbrio entre Precision e Recall durante o processo de validação.

```python
# Exibição da melhor configuração encontrada
print("Melhores parâmetros:")
print(grid_search.best_params_)

print("\nMelhor F1 na validação cruzada:")
print(grid_search.best_score_)
```

```python
# Seleção do melhor modelo encontrado pelo GridSearchCV
best_model = grid_search.best_estimator_

# Realização das previsões utilizando o modelo otimizado
y_pred_best = best_model.predict(X_test)
```

```python
# Cálculo das métricas de desempenho do modelo otimizado
best_accuracy = accuracy_score(y_test, y_pred_best)
best_precision = precision_score(y_test, y_pred_best)
best_recall = recall_score(y_test, y_pred_best)
best_f1 = f1_score(y_test, y_pred_best)

print("Accuracy :", best_accuracy)
print("Precision:", best_precision)
print("Recall   :", best_recall)
print("F1-score :", best_f1)
```

## Análise da Otimização

O GridSearchCV selecionou a melhor combinação de hiperparâmetros para o modelo Gradient Boosting.

Após a avaliação no conjunto de teste, observou-se desempenho semelhante ao modelo utilizado antes da otimização.

Esse resultado indica que a configuração inicial já apresentava boa capacidade de generalização para o conjunto de dados analisado.

Ainda assim, a etapa de otimização foi importante para validar a escolha do modelo final e atender às boas práticas de desenvolvimento de modelos de Machine Learning.

# 9. Avaliação dos Resultados

Nesta etapa, o modelo final foi avaliado utilizando Matriz de Confusão e ROC-AUC.

Essas métricas complementam a análise realizada anteriormente e permitem avaliar de forma mais detalhada a capacidade do modelo em identificar falhas operacionais.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Construção da matriz de confusão do modelo final
cm = confusion_matrix(y_test, y_pred_best)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm
)

disp.plot()

plt.title("Matriz de Confusão - Modelo Final")
plt.show()
```

## Interpretação da Matriz de Confusão

A matriz de confusão mostra que o modelo classificou corretamente a maior parte dos registros sem falha e identificou uma parcela significativa dos eventos de falha.

Os resultados indicam boa capacidade de discriminação entre as classes, com baixa quantidade de falsos positivos e falsos negativos.

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

# Cálculo das probabilidades previstas para a classe positiva
y_prob = best_model.predict_proba(X_test)[:, 1]

# Cálculo da métrica ROC-AUC
auc = roc_auc_score(y_test, y_prob)

print("ROC-AUC:", auc)
```

```python
# Construção da Curva ROC do modelo final
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(6,4))

plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.plot([0,1],[0,1],"--")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Curva ROC")
plt.legend()

plt.show()
```

## Interpretação da Curva ROC

O valor de ROC-AUC obtido foi aproximadamente 0,969.

Esse resultado indica excelente capacidade de separação entre registros com falha e sem falha.

Quanto mais próximo de 1,0 for o valor de ROC-AUC, melhor é a capacidade discriminatória do modelo. Dessa forma, o resultado obtido demonstra elevado potencial do modelo para aplicações de manutenção preditiva.

## Discussão sobre Overfitting e Underfitting

Não foram observados indícios relevantes de underfitting, uma vez que o modelo final apresentou desempenho satisfatório nas métricas de avaliação utilizadas.

Também não foram identificados indícios relevantes de overfitting, considerando o desempenho obtido no conjunto de teste e durante a validação cruzada realizada pelo GridSearchCV.

Os resultados sugerem que o modelo apresentou capacidade adequada de generalização para os dados analisados.

Entretanto, uma análise mais aprofundada poderia incluir a comparação direta entre métricas de treinamento, validação e teste, permitindo uma avaliação mais robusta do comportamento do modelo.
